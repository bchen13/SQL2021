Related to Primary 1: All the categories for the videos in our DB and which categories are popular and not.  Assuming a category with 10 or more videos is popular. And less than 10 is not popular using Case.

Question (Related to Primary 1): What are all the categories for the videos in our DB and which categories are popular and which aren't?  Assume a category with 10 or more videos is popular, videos with 10 are somewhat popular, and anything less than 10 is not popular using Case statement.

Business Justification: We want to see which categories are trending and possibly push more ads on this videos to generate revenue.  We would like to see the categories with the least amount of videos and strategize how to get more videos from their respective categories to get in this top 200.  

SQL feature(s) used: CASE, JOIN, GROUP BY, ORDER BY




```python
%%sql 
SELECT snippet_categoryId, COUNT(site_id), sc.title AS category_name, 
CASE 
        WHEN COUNT(site_id) > 10 THEN 'Category is Popular'
        WHEN COUNT(site_id) = 10 THEN 'Category is Somewhat Popular'
        ELSE 'The category is not popular'
END AS video_popularity
    FROM site_table st 
    JOIN site_category sc 
        ON sc.categoryid = st.snippet_categoryId 
    GROUP BY snippet_categoryId
    ORDER BY COUNT(site_id) DESC;
```

Question (Related to Primary 2):  What are the popular videos in each category and their ranks?

Business Justification: We want to see which videos in their categories are popular so we can possibly examine further the qualities of the video.  We can look into their thumbnail and actual content of their channel as well.  This can lead to us pushing these videos more on people's home page.  

SQL feature(s) used: JOIN, GROUP BY, VIEW, WINDOW FUNCTION

(Included the CREATE VIEW before the Query because we reference the view in the Query). 



```python
%%sql
CREATE OR REPLACE VIEW vwsite_with_category_name AS
SELECT sc.title AS category_name, 
    st.site_id, 
    st.snippet_title,
    st.snippet_description, 
    st.snippet_categoryId, 
    st.snippet_publishedAt,
    st.contentDetails_duration 
FROM site_table st
    JOIN site_category sc
        ON sc.categoryid = st.snippet_categoryId;
```

```python
 %%sql
SELECT vwst.category_name, vwst.snippet_title, ss.statistics_likeCount,snippet_categoryId,
DENSE_RANK() OVER 
    (
        PARTITION BY vwst.category_name
        ORDER BY ss.statistics_likeCount DESC
    ) AS videos_likes_rank
FROM site_stats ss 
JOIN vwsite_with_category_name vwst
    ON ss.site_id = vwst.site_id ;
```

Question (Primary):  What videos are in the top 5 most liked in their category, while the number of videos in their respective categories is greater than 30?

Business Justification: We want to see the top 5 videos most liked in their category and the categories with videos greater than 30 videos because we put in more resources such as funding, brand deals with certain channels, and ads towards certain categories, videos, and channels.  This can help us increase our ROI, save money, and time.  

SQL feature(s) used:  CTE, GROUP BY, WINDOW FUNCTION, SUBQUERY, JOIN, VIEW 


```python
%%sql

WITH videos_ranked_by_likes AS (
SELECT vwst.category_name, vwst.snippet_title, ss.statistics_likeCount,snippet_categoryId,
DENSE_RANK() OVER 
    (
        PARTITION BY vwst.category_name
        ORDER BY ss.statistics_likeCount DESC
    ) AS videos_likes_rank
FROM site_stats ss 
JOIN vwsite_with_category_name vwst
    ON ss.site_id = vwst.site_id 
)
SELECT *
FROM videos_ranked_by_likes
WHERE videos_likes_rank <= 5
AND snippet_categoryId IN 
    (
        SELECT snippet_categoryId
        FROM site_table st 
            JOIN site_category sc 
                ON sc.categoryid = st.snippet_categoryId 
        GROUP BY snippet_categoryId
        HAVING COUNT(site_id)>30 
    );
```
